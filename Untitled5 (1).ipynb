```python
import tkinter as tk
from tkinter import messagebox
from collections import defaultdict
import networkx as nx
import matplotlib.pyplot as plt

class GraphRepresentationApp:
    def __init__(self, master):
        # Initialize the GUI
        self.master = master
        self.master.title("Weighted Undirected Graph Representation App")
        self.nodes = []  # List to store nodes
        self.graph = defaultdict(dict)  # Dictionary of dictionaries for weighted edges

        # GUI elements for adding vertices and edges
        tk.Label(master, text="Vertex Name:").pack()
        self.vertex_name_entry = tk.Entry(master)
        self.vertex_name_entry.pack()

        tk.Button(master, text="Add Vertex", command=self.add_vertex).pack()

        tk.Label(master, text="Start Vertex:").pack()
        self.start_vertex_entry = tk.Entry(master)
        self.start_vertex_entry.pack()

        tk.Label(master, text="End Vertex:").pack()
        self.end_vertex_entry = tk.Entry(master)
        self.end_vertex_entry.pack()

        tk.Label(master, text="Edge Weight:").pack()
        self.edge_weight_entry = tk.Entry(master)
        self.edge_weight_entry.pack()

        tk.Button(master, text="Add Edge", command=self.add_edge).pack()
        tk.Button(master, text="Find Disconnected Graphs and Apply Prim's Algorithm", command=self.find_disconnected_graphs).pack()

    def add_vertex(self):
        # Function to add a vertex to the graph
        vertex_name = self.vertex_name_entry.get()
        if vertex_name:
            if vertex_name not in self.nodes:
                self.nodes.append(vertex_name)
                self.graph[vertex_name] = {}  # Initialize the node in the graph as an empty dictionary
                messagebox.showinfo("Info", f"Vertex {vertex_name} added.")
                self.visualize_graph()  # Visualize the graph after adding a vertex
            else:
                messagebox.showinfo("Info", "Vertex already exists!")
        else:
            messagebox.showerror("Error", "Vertex name cannot be empty!")

    def add_edge(self):
        # Function to add an edge to the graph
        start_vertex = self.start_vertex_entry.get()
        end_vertex = self.end_vertex_entry.get()
        edge_weight = self.edge_weight_entry.get()

        if start_vertex and end_vertex and edge_weight:
            try:
                edge_weight = float(edge_weight)
                if start_vertex in self.nodes and end_vertex in self.nodes:
                    if start_vertex in self.graph and end_vertex in self.graph[start_vertex]:
                        messagebox.showinfo("Info", f"Edge {start_vertex}-{end_vertex} already exists!")
                    else:
                        self.graph[start_vertex][end_vertex] = {'weight': edge_weight}  # Set the weight attribute
                        self.graph[end_vertex][start_vertex] = {'weight': edge_weight}  # Set the weight attribute for undirected graph
                        messagebox.showinfo("Info", f"Edge {start_vertex}-{end_vertex} added with weight {edge_weight}")
                        self.visualize_graph()  # Visualize the graph after adding an edge
                else:
                    messagebox.showerror("Error", "Invalid vertices! Both vertices should exist in the graph.")
            except ValueError:
                messagebox.showerror("Error", "Invalid edge weight. Please enter a valid number.")

    def visualize_graph(self):
        # Function to visualize the graph using networkx and matplotlib
        if self.graph:
            G = nx.Graph(self.graph)  # Create an undirected graph
            pos = nx.spring_layout(G)
            edge_labels = {(u, v): G[u][v]['weight'] for u, v in G.edges()}
            edge_positions = [(pos[u] + pos[v]) / 2 for u, v in G.edges()]

            plt.figure(figsize=(8, 6))
            nx.draw(G, pos, with_labels=True, node_size=2000, node_color='skyblue', font_size=20, font_weight='bold', arrows=False)

            for (x, y), weight in zip(edge_positions, edge_labels.values()):
                plt.text(x, y, str(weight), fontsize=12, ha='center')

            plt.title("Weighted Undirected Graph Representation")
            plt.show()
        else:
            messagebox.showinfo("Info", "Graph is empty!")

    def find_disconnected_graphs(self):
        # Function to find disconnected components and apply Prim's algorithm
        if self.graph:
            G = nx.Graph(self.graph)  # Create an undirected graph
            connected_components = list(nx.connected_components(G))

            if len(connected_components) > 1:
                messagebox.showinfo("Info", f"Graph has {len(connected_components)} disconnected components.")
            else:
                messagebox.showinfo("Info", "Graph is connected.")
            
            # Apply Prim's algorithm on each connected component
            for i, component in enumerate(connected_components, start=1):
                subgraph = G.subgraph(component)
                self.apply_prims_algorithm(subgraph, f"Connected Component {i}")

        else:
            messagebox.showinfo("Info", "Graph is empty!")

    def apply_prims_algorithm(self, graph, title):
        # Function to apply Prim's algorithm and visualize the minimum spanning tree
        sorted_vertices = sorted(graph.nodes())
        min_spanning_tree = nx.minimum_spanning_tree(graph)
        pos = nx.spring_layout(graph)
        edge_labels = {(u, v): graph[u][v]['weight'] for u, v in min_spanning_tree.edges()}
        edge_positions = [(pos[u] + pos[v]) / 2 for u, v in min_spanning_tree.edges()]

        plt.figure(figsize=(8, 6))
        nx.draw(min_spanning_tree, pos, with_labels=True, node_size=2000, node_color='skyblue', font_size=20, font_weight='bold', arrows=False)

        for (x, y), weight in zip(edge_positions, edge_labels.values()):
            plt.text(x, y, str(weight), fontsize=12, ha='center')

        plt.title(f"Minimum Spanning Tree - {title}")
        plt.show()

# Create the Tkinter window
root = tk.Tk()
app = GraphRepresentationApp(root)
root.mainloop()
```
